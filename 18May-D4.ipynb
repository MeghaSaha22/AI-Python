# SVM- Support Vector Machine


```python
Z = X2 + Y -> Kernel
1)Linear
2)RBF
3)Poly
4)Sigmoid
```

```python
import pandas as pd
df=pd.read_csv("Social_Network_Ads.csv")
print(df)
```

```python
x=df.iloc[:,[2,3]].values
print(x)
y=df.iloc[:,4].values
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=1)
print(x_train)
print(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
x_train=sc.fit_transform(x_train)
print(x_train)
x_test=sc.transform(x_test)
print(x_test)
```

```python
from sklearn.svm import SVC
sv=SVC(kernel='linear',random_state=1)
sv.fit(x_train,y_train)
y_pred=sv.predict(x_test)
print(y_pred)
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```python
from sklearn.svm import SVC
sv=SVC(kernel='rbf',random_state=1)
sv.fit(x_train,y_train)
y_pred=sv.predict(x_test)

from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```python
from sklearn.svm import SVC
sv=SVC(kernel='poly',random_state=1)
sv.fit(x_train,y_train)
y_pred=sv.predict(x_test)

from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```python
from sklearn.svm import SVC
sv=SVC(kernel='sigmoid',random_state=1)
sv.fit(x_train,y_train)
y_pred=sv.predict(x_test)

from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(acc)
```

# NAIVE BAYES


```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
y_pred=nb.predict(x_test)

from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(acc)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='gini',random_state=0)
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)

from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='entropy',random_state=0)
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)

from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=20,criterion='gini',random_state=0)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)

from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(acc) 
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=20,criterion='entropy',random_state=0)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)

from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(acc)
```

# Clustering
ELBOW METHOD

```python
import pandas as pd
df=pd.read_csv("Mall_Customers.csv")
print(df)
```

```python
x=df.iloc[:,[3,4]].values
print(x)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,26):
    km=KMeans(n_clusters=i,init='k-means++',random_state=0)
    km.fit(x)
    inr=km.inertia_
    wcss.append(inr)
    
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,26),wcss)
plt.xlabel("no. of cluster")
plt.ylabel("inertia")
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=5,init='k-means++',random_state=1)
y_pred=km.fit_predict(x) 
print(y_pred)
print(len(y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=10,c='red',label="cluster 0")
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=10,c='green',label="cluster 1")
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=10,c='yellow',label="cluster 2")
plt.scatter(x[y_pred==3,0],x[y_pred==3,1],s=10,c='black',label="cluster 3")
plt.scatter(x[y_pred==4,0],x[y_pred==4,1],s=10,c='purple',label="cluster 4")
```

```python
km.cluster_centers_
```

```python
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=130,marker='^',c='purple')
plt.xlabel("income")
plt.ylabel("score")
plt.show()
```
